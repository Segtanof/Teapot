```python
#import data analysis packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import json
```

```python
#open json file as dataframe
df = pd.read_json("result.json")
df
```

```python
df["ref_count"] = df["ref_task"].apply(lambda x : len(x))

df["gen_count"] = count = df["parsed_response"].apply(lambda x : len(x))
# if the number of gen_count is smaller than 80% of ref_count, drop that row
df_clean = df[df["gen_count"] >= 0.8 * df["ref_count"]]
df_clean
```

```python
# get mean similarity score base on industry
mean_score = df_clean.groupby("ind")["score"].mean()

#plot the series using column chat
fig = sns.barplot(mean_score)
fig
```

```python
x = 3
matrix = np.array(df_clean["matrix"][x])
order = df_clean["gen_order"][x]

rearranged_matrix = matrix[:, order]
sns.heatmap(rearranged_matrix)
```
